###  The dataset is highly unbalance:
0 - Normal Transaction
1 - Fraudulent Transaction 

```python
# separating the dataset for analysis 
legit_trans = credit_df[credit_df.Class==0]
fraud_trans = credit_df[credit_df.Class == 1]
```

```python
print(legit_trans.shape)
print(fraud_trans.shape)
```

# under sampling
Handling the unbalanced dataset by building sample dataset containing similar distriof normal transaction and fraudulent transaction


```python
legit_sample = legit_trans.sample(n=492)
```

```python
# concatenate the two samples
df = pd.concat([legit_sample, fraud_trans], axis=0)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Class'].value_counts()
```

```python
df.groupby('Class').mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x = df.drop(columns='Class', axis=1)
y = df['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
x_train.shape
```

```python
y_train.shape
```

# Model Iraining

```python
model = LogisticRegression()
```

```python
# training the model
model.fit(x_train, y_train)
```

# Model Evaluation
- ## Accuracy score

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print(training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(test_data_accuracy)
```
